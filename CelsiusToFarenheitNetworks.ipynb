```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
celsius_q = np.array([-40, -10, 0 , 8 , 15, 22, 38], dtype=float)
fahrenheit_a = np.array([-40, 14, 32, 46.4, 59, 71.6, 100], dtype=float)
```

```python
# We define a single layer named 'l0'
l0 = tf.keras.layers.Dense(units=1)

# We assemble it into a sequential model, explicitly defining the input shape
model = tf.keras.Sequential([tf.keras.Input(shape=(1,)), l0])
```

```python
model.compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.Adam(0.1))
```

```python
history_single_layer = model.fit(celsius_q, fahrenheit_a, epochs=500, verbose=False)
print("Finished training the model")
```

```python
import matplotlib.pyplot as plt

plt.xlabel('Epoch Number')
plt.ylabel("Loss Magnitude")
plt.plot(history.history['loss'])
plt.show()
```

```python
print("Based on the model: 100 degrees celsius is equal to ",model.predict(np.array([100.0]) )," degrees Farenheit.")
```

```python
# More Complex Research
l0 = tf.keras.layers.Dense(units=4)
l1 = tf.keras.layers.Dense(units=4)
l2 = tf.keras.layers.Dense(units=1)
```

```python
model = tf.keras.Sequential([tf.keras.Input(shape=(1,)), l0, l1, l2])

# Compile the model with the same loss and optimizer
model.compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.Adam(0.1))

# Fit the model using the existing training data
history = model.fit(celsius_q, fahrenheit_a, epochs=500, verbose=False)
print("Finished training the multi-layer model")

print("Model predicts that 100 degrees Celsius is: ", model.predict(np.array([100.0]) ))
print("These are the l0 variables: {}".format(l0.get_weights()))
print("These are the l1 variables: {}".format(l1.get_weights()))
print("These are the l2 variables: {}".format(l2.get_weights()))
```

### Interpreting Multi-Layer Model Weights

1.  **Challenge in Interpretation**: In contrast to the simple single-layer model, interpreting the weights of the more complex multi-layer model becomes significantly more challenging.

2.  **Hidden Layers and Feature Abstraction**: The multi-layer model introduces 'hidden layers' (like `l0` and `l1` in our example). These layers do not directly output the final result but instead learn and transform the input data into more abstract, internal representations or features. For instance, `l0` might capture some initial transformed representation of the Celsius input, and `l1` further refines these representations. The weights within these hidden layers are not directly mapping Celsius to Fahrenheit in a human-interpretable way; instead, they are creating intermediate 'features' that the subsequent layers use.

3.  **Contrast with Single-Layer Model**: In the single-layer model, we had a direct relationship: `Fahrenheit = weight * Celsius + bias`. The `weight` directly represented the slope, and the `bias` represented the y-intercept of this linear transformation. We could assign a clear physical meaning to these values. For example, the `l0.get_weights()` output `[array([[1.8188686]], dtype=float32), array([29.309208], dtype=float32)]` clearly suggests a slope of approximately 1.8 and an intercept of approximately 29.3.

4.  **Complex Non-Linear Transformation**: In a multi-layer model, the weights in hidden layers (e.g., `l0` and `l1`) are part of a complex, non-linear chain of calculations across multiple layers. Each weight contributes to transforming the input through several stages, and its individual value no longer directly corresponds to a simple slope or intercept of the overall input-output relationship. The output of one layer becomes the input to the next, creating a hierarchical set of transformations that are very difficult for humans to deconstruct into meaningful individual components.

5.  **Benefit of Complexity**: While this complexity makes direct interpretation difficult, it is precisely what allows multi-layer models to learn and approximate more intricate and non-linear relationships in data. This increased capacity for learning complex patterns is why multi-layer neural networks are so powerful in tasks like image recognition, natural language processing, and other areas where simple linear models fall short. The model learns to build up complex features from simpler ones without us explicitly telling it what those features should be.
